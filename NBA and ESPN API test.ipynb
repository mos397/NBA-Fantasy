```python
members_json = espn_raw_data['members']
members_df = pd.json_normalize(members_json)
```

```python
members_df.head()
```

```python
teams_json = espn_raw_data['teams']
teams_df = pd.json_normalize(teams_json)
```

```python
teams_df.head()
```

```python
# choose week to get game numbers for home and away, weeknumber*5-5 and weeknumber*5-4
rosters_df = pd.DataFrame()

def gameids(weeknumber):
    start = (weeknumber*5)-5
    end = (weeknumber*5)-1
    return range(start, end)

current_weekid = espn_raw_data['gameId']

for games in gameids(current_weekid):
    away_df = pd.json_normalize(espn_raw_data['schedule'][games]['away']['rosterForCurrentScoringPeriod']['entries'])
    home_df = pd.json_normalize(espn_raw_data['schedule'][games]['home']['rosterForCurrentScoringPeriod']['entries'])
    rosters_df = pd.concat([away_df,home_df,rosters_df], ignore_index=True)
```

```python
combined_df = teams_df.merge(rosters_df, left_on='id', right_on='playerPoolEntry.onTeamId')
```
